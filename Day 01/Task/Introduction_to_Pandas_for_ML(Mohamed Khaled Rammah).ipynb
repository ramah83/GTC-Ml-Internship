<h4 style="color:#6c0e0eff">1. Importing Essential Python Libraries for Data Analysis and Visualization</h4>

<h4 style="color:#6c0e0eff">2. Loading the Iris Dataset into a Pandas DataFrame</h4>

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
df = pd.DataFrame(data=iris_data['data'], columns=iris_data['feature_names'])

print("Type of the object:", type(df))
print("")
print("Shape of the DataFrame (Rows, Columns):", df.shape)
```

<h4 style="color:#6c0e0eff">8. Selecting Multiple Columns from the DataFrame and Displaying the First Five Rows of the DataFrame </h4>


```python
sepal_data = df[['sepal length (cm)', 'sepal width (cm)']]
sepal_data.head()
```

<h4 style="color:#6c0e0eff">9. Visualizing the Distribution of Sepal Lengths with a Histogram</h4>

```python
import matplotlib.pyplot as plt
plt.hist(df['sepal length (cm)'], color='skyblue', edgecolor='black', bins=15)
plt.title('Distribution of Sepal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

<h4 style="color:#6c0e0eff">10. Visualizing the Distribution of Sepal Lengths with Seaborn Histogram and KDE</h4>

```python
import seaborn as sns
plt.figure(figsize=(8, 5)) 
sns.histplot(df['sepal length (cm)'], kde=True, color='skyblue', bins=15)
plt.title('Distribution of Sepal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

<h4 style="color:#6c0e0eff">11. Visualizing Feature Correlations with a Heatmap</h4>

```python
plt.figure(figsize=(8, 5)) 
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', cbar=False)
plt.title('Correlation Heatmap',y=1.05)
plt.show()
```

<h4 style="color:#6c0e0eff">12. Comparing Feature Distributions by Species with Boxplots</h4>

```python
if 'species' not in df.columns:
    df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
fig, axes = plt.subplots(2, 2, figsize=(10,8))
features = iris_data.feature_names
for ax, feat in zip(axes.ravel(), features):
    sns.boxplot(data=df, x='species', y=feat, ax=ax)
    ax.set_title(f'Boxplot of {feat} by Species')
plt.tight_layout()
plt.show()
```
